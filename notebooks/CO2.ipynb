```python
import pandas as pd
pd.set_option('display.float_format', '{:.2f}'.format)
from sqlalchemy import create_engine
```

```python
engine = create_engine(
    "mysql+pymysql://root@localhost:3306/climate_watch"
)
```

```python
engine.connect()
```

```python
CO2 = pd.read_sql(
    "SELECT * FROM co2",
    engine
)
```

```python
CO2[['CO2']].describe()
```

```python
landenperjaar = CO2.groupby('Year').count()
print(landenperjaar)
```

```python
laatstejaar = CO2['Year'].max()
print(laatstejaar)
```

```python
eu_countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus',
    'Czechia', 'Denmark', 'Estonia', 'Finland',
    'France and Monaco', 'Germany', 'Greece', 'Hungary', 'Ireland',
    'Italy, San Marino and the Holy See', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta',
    'Netherlands', 'Poland', 'Portugal', 'Romania',
    'Slovakia', 'Slovenia', 'Spain', 'Sweden'
]  
```

```python
eu_per_year = (CO2[CO2['Country'].isin(eu_countries)].groupby('Year')['CO2'].sum().reset_index())
eu_per_year['Country'] = 'European Union'
print(eu_per_year)
```

```python
CO2_zonder_eu = CO2[~CO2['Country'].isin(eu_countries)]
```

```python
import pandas as pd

CO2_nieuw = pd.concat([CO2_zonder_eu, eu_per_year], ignore_index=True)
aantallanden = CO2_nieuw.groupby('Year').count()
print(aantallanden)
```

```python
laatstejaarco2 = CO2_nieuw[CO2_nieuw['Year'] == laatstejaar]
print(laatstejaarco2)
```

```python
top5 = laatstejaarco2.sort_values(by='CO2', ascending=False).head(5)
print(top5)
```

```python
totaal_co2 = laatstejaarco2['CO2'].sum()
print(totaal_co2)
```

```python
top5_co2 = top5['CO2'].sum()
print(top5_co2)
```

```python
rest_van_de_wereld_co2 = totaal_co2 - top5_co2
print(rest_van_de_wereld_co2)
```

```python
eerstejaar = CO2_nieuw['Year'].min()
print(eerstejaar)
```

```python
eerstejaarCO2 = CO2_nieuw[CO2_nieuw['Year'] == eerstejaar]
print(eerstejaarCO2)
```

```python
print(laatstejaarco2)
```

```python
verandering_in_co2 = ((laatstejaarco2.groupby("Country")['CO2'].sum() - eerstejaarCO2.groupby("Country")['CO2'].sum()) / eerstejaarCO2.groupby("Country")['CO2'].sum()) * 100
print(verandering_in_co2)
```

```python
top5_stijging = verandering_in_co2.sort_values(ascending=False).head(5)
print(top5_stijging)
```

```python
top5_dalen = verandering_in_co2.sort_values(ascending=True).head(5)
print(top5_dalen)
```

```python
wereld_verandering_in_co2 = ((laatstejaarco2['CO2'].sum() - eerstejaarCO2['CO2'].sum()) / eerstejaarCO2['CO2'].sum()) * 100
print(wereld_verandering_in_co2)
```

```python
totaal = top5['CO2'].sum() + rest_van_de_wereld_co2
pie_chart = pygal.Pie(background='white')
pie_chart.title = 'Top 5 CO2 Emitting Countries in ' + str(laatstejaar)
pie_chart.value_formatter = lambda x: f'{(x / totaal) * 100:.1f}%'


pie_chart.print_values = True
pie_chart.data_labels = True

for _, row in top5.iterrows():
    pie_chart.add(row['Country'], row['CO2'])

pie_chart.add('Rest of the World', rest_van_de_wereld_co2)

pie_chart.render_to_file('top5_co2_emissions.svg')
```

```python

top5_over_tijd = CO2_nieuw[
    CO2_nieuw['Country'].isin(top5['Country'])
].pivot(index='Year', columns='Country', values='CO2')

print(top5_over_tijd)

```

```python
line_chart = pygal.Line(
    style=pygal.style.Style(background='white', plot_background='white'))
line_chart.title = 'CO2 Emissions of Top 5 Countries Over Time'


years = top5_over_tijd.index.tolist()

line_chart.x_labels = [
    str(year) if year % 10 == 0 else '' for year in years
]
line_chart.x_label_rotation = 25

line_chart.y_title = 'CO2 Emissions (Million Tonnes)'
line_chart.x_title = 'Year'

for country in top5_over_tijd.columns:
    line_chart.add(country, top5_over_tijd[country].tolist())

line_chart.render_to_file('top5_stijgers_co2_over_time.svg')
```
